## 1. Build a DataFrame & inspect it
`head` / `info` / `describe` are the first things to run on any dataset.

```python
df = pd.DataFrame({
    'name':  ['Ada', 'Alan', 'Grace', 'Linus', 'Katherine', 'Guido'],
    'team':  ['A', 'B', 'A', 'B', 'A', 'B'],
    'score': [91, 78, 88, 64, 95, 72],
    'hours': [6.0, 3.5, 5.0, 2.0, 6.5, 3.0],
})
print(df.shape)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## 2. Select columns and rows
`.loc` selects by **label**, `.iloc` by **integer position**.

```python
print(df['score'].head())          # one column = a Series
print(df[['name', 'score']].head())# several columns
print(df.loc[0])                   # row with index label 0
print(df.iloc[0:3, 0:2])           # first 3 rows, first 2 cols
```

## 3. Filter rows with boolean conditions

```python
print(df[df['score'] >= 85])
print(df[(df['score'] >= 85) & (df['hours'] < 6)])
```

## 4. Create new columns (vectorized, like NumPy)

```python
df['passed'] = df['score'] >= 80
df['per_hour'] = (df['score'] / df['hours']).round(2)
df
```

## 5. Group & aggregate — 'per-category' questions

```python
print(df.groupby('team')['score'].mean())
print(df.groupby('passed')['score'].agg(['mean', 'size']))
print(df['team'].value_counts())
```

## 6. Missing data
Real data has holes (`NaN`). Find them, then decide what to do.

```python
d2 = df.copy()
d2.loc[1, 'score'] = np.nan        # poke a hole
print('missing per column:\n', d2.isna().sum())
print('filled with mean:\n',
      d2['score'].fillna(d2['score'].mean()).round(1).tolist())
```

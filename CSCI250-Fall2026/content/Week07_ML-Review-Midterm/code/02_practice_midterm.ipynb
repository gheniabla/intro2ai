## Part D — Compute by hand, then verify
**D1.** A classifier gives `TP=30, FP=20, FN=10, TN=40`. Compute **accuracy, precision, recall, F1** by hand, then check with the cell below.

```python
TP, FP, FN, TN = 30, 20, 10, 40
# Fill in, then run to check your hand calculation:
accuracy  = (TP + TN) / (TP + FP + FN + TN)
precision = TP / (TP + FP)
recall    = TP / (TP + FN)
f1 = 2 * precision * recall / (precision + recall)
print(round(accuracy,2), round(precision,2), round(recall,2), round(f1,2))
```

**D2.** Gradient descent minimizes `f(w) = (w-5)**2` with `w0=0`, `lr=0.1`. What is `grad` at the first step, and is the first update toward or away from 5? (grad = 2*(w-5).) Verify below.

```python
w, lr = 0.0, 0.1
grad = 2 * (w - 5)
w_next = w - lr * grad
print('first grad =', grad, ' w after one step =', w_next)
```

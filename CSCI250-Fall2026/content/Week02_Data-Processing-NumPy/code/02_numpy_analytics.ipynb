## 1. Make a tiny dataset (so the notebook is self-contained)
Columns: **hours_studied, prior_gpa, attendance, exam_score**.

```python
csv_text = '''hours_studied,prior_gpa,attendance,exam_score
2.0,2.8,0.60,55
4.5,3.1,0.85,72
1.0,2.5,0.40,48
6.0,3.6,0.95,88
3.0,3.0,0.75,65
5.5,3.8,0.90,91
0.5,2.2,0.30,40
4.0,3.3,0.80,78'''
with open('students.csv', 'w') as f:
    f.write(csv_text)
print('wrote students.csv')
```

## 2. Load with `np.genfromtxt`
We skip the header row and read the four numeric columns into a 2-D array.

```python
data = np.genfromtxt('students.csv', delimiter=',', skip_header=1)
cols = ['hours_studied', 'prior_gpa', 'attendance', 'exam_score']
print('shape:', data.shape)
print(data)
```

## 3. Per-column means (axis aggregation)

```python
means = data.mean(axis=0)
for name, mu in zip(cols, means):
    print(f'{name:14s} mean = {mu:.3f}')
```

## 4. Filter rows by a condition (boolean mask)
Which students scored above the class average exam score? No loops.

```python
exam = data[:, 3]
above = data[exam > exam.mean()]
print('rows above average exam score:')
print(above)
```

## 5. Find the top student (argmax)

```python
best_idx = data[:, 3].argmax()
print('top student row index:', best_idx)
print('their record         :', data[best_idx])
```

## 6. Standardize the features (broadcasting)
Subtract each column's mean and divide by its std — what models usually want.

```python
X = data[:, :3]                      # feature columns
X_std = (X - X.mean(axis=0)) / X.std(axis=0)
print('standardized feature means (~0):', X_std.mean(axis=0).round(3))
print('standardized feature stds  (~1):', X_std.std(axis=0).round(3))
```

## 7. Correlation: does studying track the exam score?
A quick `np.corrcoef` between hours_studied and exam_score.

```python
r = np.corrcoef(data[:, 0], data[:, 3])[0, 1]
print(f'corr(hours_studied, exam_score) = {r:.3f}')
```

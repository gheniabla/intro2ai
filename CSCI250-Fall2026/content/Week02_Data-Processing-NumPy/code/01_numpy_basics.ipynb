## 1. Creating arrays

```python
a = np.array([[1, 2, 3], [4, 5, 6]])
print(a)
print('zeros:\n', np.zeros((2, 3)))
print('arange:', np.arange(0, 10, 2))
print('linspace:', np.linspace(0, 1, 5))
```

## 2. Inspecting an array
Every array has `shape`, `ndim`, `dtype`, `size` — check these constantly.

```python
print('shape', a.shape)
print('ndim ', a.ndim)
print('dtype', a.dtype)
print('size ', a.size)
```

## 3. Reshaping
The same data, viewed in a new shape. `-1` means 'infer this dimension'.

```python
v = np.arange(12)
print(v.reshape(3, 4))
print('flatten back:', v.reshape(3, 4).reshape(-1))
```

## 4. Vectorization — and why it is faster
Element-wise math runs in C. Use `%timeit` to *see* the speedup.

```python
N = 1_000_000
py = %timeit -o -q [x * 2 for x in range(N)]
nm = %timeit -o -q np.arange(N) * 2
print(f'python loop: {py.best:.4f}s   numpy: {nm.best:.6f}s')
```

```python
a = np.array([1.0, 4.0, 9.0])
print('sqrt :', np.sqrt(a))
print('>3   :', a > 3)
print('*10+1:', a * 10 + 1)
```

## 5. Broadcasting
Combine different shapes with no loop — e.g. add a row to every row.

```python
matrix = np.ones((3, 3))
row = np.array([1, 2, 3])
print(matrix + row)         # row added to each row
prices = np.array([10., 20., 30.])
print('with tax:', prices * 1.08)
```

## 6. Indexing, slicing & boolean masks

```python
m = np.arange(12).reshape(3, 4)
print('m[1,2]   =', m[1, 2])
print('col 0    =', m[:, 0])
print('subblock =\n', m[0:2, 1:3])

data = np.array([3, -1, 7, -5, 9])
print('positives:', data[data > 0])
data[data < 0] = 0
print('clipped  :', data)
```

## 7. Axis operations
`axis=0` collapses rows (per-column result); `axis=1` collapses columns (per-row).

```python
scores = np.array([[90, 80, 70],
                   [60, 75, 95]])
print('per subject (axis=0):', scores.mean(axis=0))
print('per student (axis=1):', scores.mean(axis=1))
print('best subject idx    :', scores.argmax(axis=1))
```

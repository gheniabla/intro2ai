## 1. Data types

```python
scalars = (42, 3.14, True, 'ai')
nums = [1, 2, 3]            # list (mutable)
point = (3, 4)              # tuple (immutable)
config = {'model': 'claude-sonnet-4-6', 'temp': 0.7}  # dict
unique = {1, 2, 2, 3}       # set
print(scalars, nums, point, config, unique)
```

## 2. Control flow

```python
for i in range(5):
    print(i, 'even' if i % 2 == 0 else 'odd')
```

## 3. Functions (with type hints)

```python
def greet(name: str, excited: bool = False) -> str:
    msg = f'Hello, {name}'
    return msg + '!' if excited else msg

print(greet('CSCI 250'))
print(greet('AI', excited=True))
```

## 4. Comprehensions & f-strings

```python
squares = [x * x for x in range(10)]
evens = [x for x in squares if x % 2 == 0]
print(f'squares={squares}')
print(f'{len(evens)} even squares: {evens}')
```

## 5. Modules

```python
import math, statistics
data = [2, 4, 4, 4, 5, 5, 7, 9]
print('mean =', statistics.mean(data))
print('stdev =', round(statistics.stdev(data), 3))
print('sqrt(2) =', math.sqrt(2))
```
